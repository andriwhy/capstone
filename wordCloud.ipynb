```python
import pandas as pd
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from collections import Counter
import string

file_path = 'drive/Shareddrives/ML/Dataset/Stemmed/stemmed_train.csv'
df = pd.read_csv(file_path)

df['tokens_indonesia'] = df['text_indonesia'].apply(lambda x: x.split())

all_tokens = [token for tokens in df['tokens_indonesia'] for token in tokens]

word_counts = Counter(all_tokens)

wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(word_counts)

plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()

top_words = word_counts.most_common(10)

words, freqs = zip(*top_words)

plt.figure(figsize=(10, 5))
plt.bar(words, freqs, color='skyblue')
plt.xlabel('Kata')
plt.ylabel('Frekuensi')
plt.title('10 Kata Teratas dalam Kolom text_indonesia')
plt.xticks(rotation=45)
plt.show()
```
