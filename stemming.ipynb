```python
import io
import tensorflow as tf
import pandas as pd
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
```

```python
factory = StemmerFactory()
stemmer = factory.create_stemmer()
```

```python
df = pd.read_csv('Dataset/modified_training.csv')
```

```python
df['text_indonesia']
```

```python
def lemmatize_text(text):
    """
    Args:
        texts (list): List of text strings to be lemmatized.

    Returns:
        list: List of lemmatized text strings.
    """
    if isinstance(text, str):
        text = text.lower()
        text = stemmer.stem(text)
    return text
```

```python
df['text_indonesia'] = df['text_indonesia'].apply(lemmatize_text)
```

```python
df['text_indonesia']
```

```python
df.to_csv('Dataset/stemmed_train.csv', index=False)

print("DataFrame saved to 'Dataset/stemmed_train.csv'")
```
