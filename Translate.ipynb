```python
import pandas as pd
import csv
from googletrans import Translator
import time
```

```python
!pip install tensorflow
```

```python
input_file = 'Dataset/validation.csv'
output_file = 'Dataset/translated_validation.csv'
failed_file = 'Dataset/failed_validation.csv'
```

```python
# data = pd.read_csv(input_file, skiprows=9586, header=None, names=['text', 'label'])
data = pd.read_csv(input_file)
data
```

```python
translator = Translator()

starting = 1

def translate_text(text, retries=5, delay=2):
    for attempt in range(retries):
        try:
            translated = translator.translate(text, src='en', dest='id').text
            return translated
        except Exception as e:
            print(f"Error: {e}. Retrying ({attempt + 1}/{retries})...")
            time.sleep(delay)
    return None

for index, row in data.iterrows():
    translated_text = translate_text(row['text'])
    if translated_text is None:
        with open(failed_file, mode='a', newline='', encoding='utf-8') as file:
            writer = csv.writer(file)
            writer.writerow([row['text']])
        print(f"Baris ke-{starting + index} gagal diterjemahkan dan disimpan di {failed_file}.")
    else:
        with open(output_file, mode='a', newline='', encoding='utf-8') as file:
            writer = csv.writer(file)
            writer.writerow([row['text'], row['label'], translated_text])
        print(f"Baris ke-{starting + index} berhasil diterjemahkan dan disimpan.")
```
